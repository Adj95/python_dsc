# ENCODING NUMERICAL FEATURES -BINNING


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("/content/train (1).csv" , usecols = ['Age' , 'Fare' , 'Survived'])
df.head()
```

```python
df.dropna(inplace= True)
```

```python
x= df.iloc[: , 1:3]
y = df.iloc[: , 0]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
k_age = KBinsDiscretizer(n_bins = 10 , encode ='ordinal' , strategy ='quantile')
k_fare = KBinsDiscretizer(n_bins = 10 , encode ='ordinal' , strategy ='quantile')

trf = ColumnTransformer([
    ('first' , k_age , [0]),
    ("second" , k_fare , [1])
])
```

```python
x_train_trf = trf.fit_transform(x_train)
x_test_trf = trf.transform(x_test)
```

```python
output = pd.DataFrame({
    'age_trf' : x_train_trf[:,0],
    'age' : x_train['Age'],
    'fare_trf' : x_train_trf[:,1],
    'fare' :x_train['Fare']

})
output.head()
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train_trf , y_train)
y_pred = clf.predict(x_test_trf)
accuracy_score(y_test , y_pred)
```
