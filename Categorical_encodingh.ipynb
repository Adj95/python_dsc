## **ENCODING CATEGORIACL DATA**


```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df = pd.read_csv("/content/customer.csv")
df = df.iloc[: ,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test ,y_train,y_test = train_test_split(df.drop('purchased' , axis = 1) , df['purchased'] , test_size =0.3 , random_state=0)
x_train.shape , y_train.shape , x_test.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories =[['Poor' , 'Average' , 'Good'] , ['School' , 'UG' , 'PG']])
oe.fit(x_train)
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
x_train
```

**label** **encoding**


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)
y_train
```

# COLUMN TRANSFORMER

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder

df2 = pd.read_csv("/content/covid_toy.csv")
from sklearn.model_selection import train_test_split
x_train , x_test ,y_train,y_test = train_test_split(df2.drop('has_covid' , axis = 1) , df2['has_covid'] , test_size =0.2 , random_state=0)
x_train
```

```python
from sklearn.compose import ColumnTransformer
trans = ColumnTransformer( transformers = [
    ('tnf1' , OrdinalEncoder(categories = [['Mild' , 'Strong']]) ,['cough']),
    ('tnf2'  , SimpleImputer() , ['fever']),
    ('tnf3' , OneHotEncoder(drop = 'first') , ['gender' , 'city'])


] , remainder = 'passthrough')
x_train = trans.fit_transform(x_train)
x_test=trans.transform(x_test)
x_train
```
