# **PIPELINES**

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline , make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectKBest , chi2

```

```python
df = pd.read_csv("/content/train (1).csv")
df.head()
```

```python
df.drop(columns= ['PassengerId' , 'Name' ,'Ticket', 'Cabin'] , inplace  = True )
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns = ['Survived']) , df['Survived'] , test_size = 0.2 , random_state = 0)
x_train
```

```python
trf1 = ColumnTransformer(transformers = [
    ('impute_age' , SimpleImputer() , [2]),
    ('embarked_imputer' , SimpleImputer(strategy='most_frequent') , [6] )

] ,  remainder = 'passthrough')
```

```python
trf2 = ColumnTransformer (transformers = [
    ('ohe_sex_embarked' , OneHotEncoder(sparse_output = False ,handle_unknown = 'ignore' ) , [3,1])
] , remainder = 'passthrough')
```

```python
trf3 = ColumnTransformer(transformers = [
    ('scale' , MinMaxScaler() , slice(0,10))
] , remainder = 'passthrough')
```

```python
trf4 = SelectKBest(score_func = chi2 , k = 8)
```

```python
trf5 = DecisionTreeClassifier()
```

```python
pipe = Pipeline([
    ('trf1' , trf1),
    ('trf2' , trf2),
    ('trf3' , trf3),
    ('trf4' , trf4),
    ('trf5' , trf5)
    ])
```

```python
pipe.fit(x_train ,y_train)
```
